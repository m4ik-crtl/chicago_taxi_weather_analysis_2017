## 2 - Testando Hipóteses

#### 2.1 - Análise e verificação dos dados fornecidos

```python
# Importar o arquivo
df_trips_loop = pd.read_csv('/datasets/project_sql_result_07.csv')
df_trips_loop.info()
```

```python
# Convertendo a data para o tipo de dado correto
df_trips_loop['start_ts'] = pd.to_datetime(df_trips_loop['start_ts'])
df_trips_loop.info()
```

```python
# Criar a coluna com dia da semana
df_trips_loop['day_of_week'] = df_trips_loop['start_ts'].dt.dayofweek  # segunda=0, sábado=5
df_trips_loop.head()
```

```python
# Filtrar somente os sábados
saturdays = df_trips_loop[df_trips_loop['day_of_week']== 5]
```

```python
# Classificar o clima (tempo ruim = 'Bad')
saturdays['is_rainy'] = saturdays['weather_conditions'] == 'Bad'

# Separar os dois grupos
rainy = saturdays[saturdays['is_rainy'] == True]['duration_seconds']
not_rainy = saturdays[saturdays['is_rainy'] == False]['duration_seconds']

# Calcular variâncias
rainy_var = rainy.var()
not_rainy_var = not_rainy.var()
print(f'Variância (chuvoso): {rainy_var:.2f}')
print(f'Variância (não chuvoso): {not_rainy_var:.2f}')

# Verificar se variâncias são similares
# Regra comum: se a razão entre as variâncias for < 4, pode-se considerar equal_var=True
var_ratio = max(rainy_var, not_rainy_var) / min(rainy_var, not_rainy_var)
print(f'Razão entre variâncias: {var_ratio:.2f}')

# Escolher equal_var com base na razão
equal_var_flag = var_ratio < 4
```

#### 2.2 - Teste de hipótese

>A duração média dos passeios do Loop para o Aeroporto Internacional O'Hare muda nos sábados chuvosos

### Teste de Hipótese: Duração média das corridas do Loop até o Aeroporto O'Hare

**Hipótese nula (H₀):**  
A duração média das corridas **não muda** nos sábados chuvosos.  
Ou seja, as condições meteorológicas não afetam significativamente o tempo de viagem.

**Hipótese alternativa (H₁):**  
A duração média das corridas **muda** nos sábados chuvosos.  
Ou seja, chuva ou tempestade influenciam a duração da corrida.

> Como não estamos afirmando se a média será **maior ou menor**, este é um **teste bicaudal (two-sided)**.

---

### Nível de significância (α)

Para este teste, escolhemos um **nível de significância de 5% (α = 0.05)**.  
Esse valor é amplamente utilizado em testes estatísticos e representa uma margem de erro aceitável:

- Se **valor-p < 0.05**, rejeitamos H₀: há evidência estatística de que a duração muda com o clima.
- Se **valor-p ≥ 0.05**, não rejeitamos H₀: não há evidência suficiente para afirmar que a chuva afeta a duração.

O valor de α = 0.05 significa que aceitamos uma chance de 5% de rejeitar a hipótese nula **por engano** (erro tipo I).


```python
# Definir nível de significância
alpha = 0.05

# Realizar o teste t com base na igualdade de variâncias
results = stats.ttest_ind(rainy, not_rainy, equal_var=equal_var_flag)

# Exibir o valor-p
print('valor-p:', results.pvalue)

# Verificar resultado do teste
if results.pvalue < alpha:
    print("Rejeitamos a hipótese nula: há diferença significativa na duração média das corridas.")
else:
    print("Não rejeitamos a hipótese nula: não há diferença significativa na duração média das corridas.")
```
